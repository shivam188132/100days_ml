```python
import http.client

conn = http.client.HTTPSConnection("instagram-scraper-api2.p.rapidapi.com")

headers = {
    'x-rapidapi-key': "xxxxxxxxx",
    'x-rapidapi-host': "xxxxxxxx"
}

conn.request("GET", "/v1/likes?code_or_id_or_url=CxYQJO8xuC6", headers=headers)

res = conn.getresponse()
data = res.read()

decoded_data = data.decode("utf-8")
```

```python
import pandas as pd
import json as js
```

```python
import pandas as pd
import json

# Assuming 'decoded_data' is a JSON string
data_dict = json.loads(decoded_data)

# If the JSON has a 'data' key containing a list of records:
df = pd.DataFrame(data_dict.get('data', []))
```

```python
df.head()
```

```python
# Assuming 'df' is your DataFrame with the 'items' column containing JSON data
normalized_df = pd.json_normalize(df['items'])  # Normalize the JSON data

# Extract the desired columns
result_df = normalized_df[['full_name', 'id', 'profile_pic_id', 'profile_pic_url']]

# Display the result
result_df.to_csv('instagram_data.csv', index=False)
```
